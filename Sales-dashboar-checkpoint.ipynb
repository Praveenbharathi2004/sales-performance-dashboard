```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect("sales_db.sqlite")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS customers (
    customer_id INTEGER PRIMARY KEY,
    customer_name TEXT,
    region TEXT
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS products (
    product_id INTEGER PRIMARY KEY,
    product_name TEXT,
    category TEXT,
    price REAL
);
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    sale_id INTEGER PRIMARY KEY,
    sale_date TEXT,
    customer_id INTEGER,
    product_id INTEGER,
    quantity INTEGER,
    unit_price REAL
);
""")

conn.commit()
```

```python
sales_df = pd.read_sql("SELECT * FROM sales", conn)
products_df = pd.read_sql("SELECT * FROM products", conn)
customers_df = pd.read_sql("SELECT * FROM customers", conn)

sales_df.head(), products_df.head(), customers_df.head()
```

```python
df = (
    sales_df
    .merge(products_df, on="product_id", how="left")
    .merge(customers_df, on="customer_id", how="left")
)

df.head()
```

```python
df["sale_date"] = pd.to_datetime(df["sale_date"])
df["revenue"] = df["quantity"] * df["unit_price"]
df["month"] = df["sale_date"].dt.to_period("M").dt.to_timestamp()

df[["sale_date", "customer_name", "product_name", "quantity", "unit_price", "revenue", "month"]].head()
```

```python
total_revenue = df["revenue"].sum()

revenue_by_product = (
    df.groupby("product_name")["revenue"]
      .sum()
      .sort_values(ascending=False)
)

monthly_revenue = (
    df.groupby("month")["revenue"]
      .sum()
      .reset_index()
)

total_revenue, revenue_by_product, monthly_revenue
```
